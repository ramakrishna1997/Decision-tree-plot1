```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iri=load_iris()
```

```python
iri.keys()
```

```python
iri["target"]
```

```python
iri["feature_names"]
```

```python
iri["data"]
```

```python
iris=pd.DataFrame(iri.data,columns=iri["feature_names"])
```

```python
iris["SPECIES"]=pd.Categorical.from_codes(iri.target,iri.target_names)
```

```python
iris
```

```python
x=iris.iloc[:,:-1]
y=iris.iloc[::,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clt=DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
clt.fit(X_train,y_train)
```

```python
pred=clt.predict(X_test)
```

```python
pred
```

```python
from sklearn.tree import plot_tree
```

```python
fig=plt.figure(figsize=(15,12))
plot_tree(clt,filled=True)
plt.show()
```
